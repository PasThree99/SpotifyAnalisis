```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.cluster import KMeans as km
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = [10, 5]
```

##Limpiar los datos

```python
df = pd.read_csv('/content/drive/MyDrive/Análisis de mi spotify/spotify.csv',encoding = "ISO-8859-1")
df = df[['name','acousticness','danceability','energy','instrumentalness','liveness','speechiness','valence']]
df.set_index('name',inplace=True)
df.head()
```

##Reducción de componentes

```python

pca = PCA(n_components=3)
pca.fit(df)
CP = pca.transform(df)
CP = pd.DataFrame(CP,columns=['c1','c2','c3'])
```

##Clustering

```python
modeloKm = km (n_clusters=15)
modeloKm.fit(df)
CP['grupo'] = modeloKm.predict(df)

df['grupo'] =  modeloKm.predict(df)

CP0 = CP[ CP['grupo'] == 0]
CP1 = CP[ CP['grupo'] == 1]
CP2 = CP[ CP['grupo'] == 2]
CP3 = CP[ CP['grupo'] == 3]
CP4 = CP[ CP['grupo'] == 4]
CP5 = CP[ CP['grupo'] == 5]
CP6 = CP[ CP['grupo'] == 6]
CP7 = CP[ CP['grupo'] == 7]
CP8 = CP[ CP['grupo'] == 8]
CP9 = CP[ CP['grupo'] == 9]
CP10 = CP[ CP['grupo'] == 10]
CP11 = CP[ CP['grupo'] == 11]
CP12 = CP[ CP['grupo'] == 12]
CP13 = CP[ CP['grupo'] == 13]
CP14 = CP[ CP['grupo'] == 14]
```
